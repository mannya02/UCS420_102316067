### Q1. Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports, technology, food, books, etc.).
#### 1. Convert text to lowercase and remove punctuation using re.
#### 2. Tokenize the text into words and sentences.
#### 3. Split using split() and word_tokenize() and compare how Python split and NLTK’s word_tokenize() differ.
#### 4. Remove stopwords (using NLTK's stopwords list).
#### 5. Display word frequency distribution (excluding stopwords).

```python
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('stopwords')

text = """Generative AI is a transformative technology that enables machines to create new content such as text, images, music, and even software code. It works by learning patterns from large datasets and using that knowledge to generate original outputs. Tools like ChatGPT and DALL·E make it accessible for content creators, educators, and developers. GenAI enhances productivity and creativity, but also introduces challenges related to ethics, misinformation, and authenticity. Balancing innovation with responsibility is essential as the technology evolves."""

text_cleaned = re.sub(r'[^\w\s]', '', text.lower())
print("Cleaned Text:\n", text_cleaned, "\n")

words = word_tokenize(text_cleaned)
sentences = sent_tokenize(text)
print("Tokenized Words:\n", words)
print("\nTokenized Sentences:\n", sentences)

split_words = text_cleaned.split()
print("\nUsing split():\n", split_words)
print("\nUsing word_tokenize():\n", words)

stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w not in stop_words]
print("\nFiltered Words (No Stopwords):\n", filtered_words)

fdist = FreqDist(filtered_words)
fdist.plot(10, title="Top 10 Word Frequencies (Excluding Stopwords)")
```

### Q2. Using the same paragraph from Q1:
#### 1. Extract all words with only alphabets using re.findall()
#### 2. Remove stop words using NLTK’s stopword list
#### 3. Perform stemming with PorterStemmer
#### 4. Perform lemmatization with WordNetLemmatizer
#### 5. Compare the stemmed and lemmatized outputs and explain when you’d prefer one over the other.

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

text = """Generative AI is a transformative technology that enables machines to create new content such as text, images, music, and even software code. It works by learning patterns from large datasets and using that knowledge to generate original outputs. Tools like ChatGPT and DALL·E make it accessible for content creators, educators, and developers. GenAI enhances productivity and creativity, but also introduces challenges related to ethics, misinformation, and authenticity. Balancing innovation with responsibility is essential as the technology evolves."""

only_words = re.findall(r'\b[a-zA-Z]+\b', text.lower())
print("Words with only alphabets:\n", only_words)

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in only_words if word not in stop_words]
print("\nAfter removing stopwords:\n", filtered_words)

ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_words]
print("\nStemmed Words:\n", stemmed_words)

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("\nLemmatized Words:\n", lemmatized_words)

print("\nComparison (Word : Stemmed vs Lemmatized):")
for original, stem, lemma in zip(filtered_words, stemmed_words, lemmatized_words):
    print(f"{original} : {stem} | {lemma}")

print("\n Use Stemming when speed matters and accuracy is less important.")
print(" Use Lemmatization when you need correct base forms and grammatical meaning.")
```

### Q3. Choose 3 short texts of your own (e.g., different news headlines, product reviews).
#### 1. Use CountVectorizer to generate the Bag of Words representation.
#### 2. Use TfidfVectorizer to compute TF-IDF scores.
#### 3. Print and interpret the top 3 keywords from each text using TF-IDF.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import numpy as np
import pandas as pd

texts = [
    "Generative AI is revolutionizing content creation.",
    "Blockchain ensures data security and decentralization.",
    "AI and blockchain are shaping the future of technology."
]

count_vectorizer = CountVectorizer()
bow_matrix = count_vectorizer.fit_transform(texts)

print("=== Bag of Words (BoW) ===")
print(pd.DataFrame(bow_matrix.toarray(), columns=count_vectorizer.get_feature_names_out()))

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(texts)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

print("\n=== TF-IDF Scores ===")
print(tfidf_df.round(3))

print("\n=== Top 3 Keywords per Text (TF-IDF) ===")
for i, row in enumerate(tfidf_matrix.toarray()):
    top_indices = row.argsort()[-3:][::-1]
    top_words = [tfidf_vectorizer.get_feature_names_out()[j] for j in top_indices]
    print(f"Text {i+1}: {top_words}")
```

### Q4. Write 2 short texts (4–6 lines each) describing two different technologies (e.g., AI vs Blockchain).
#### 1. Preprocess and tokenize both texts.
#### 2. Calculate:
##### a. Jaccard Similarity using sets
##### b. Cosine Similarity using TfidfVectorizer + cosine_similarity()
##### c. Analyze which similarity metric gives better insights in your case.

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

text1 = """Artificial Intelligence (AI) enables machines to think and learn like humans.
It powers applications like voice recognition, recommendation systems, and autonomous vehicles."""

text2 = """Blockchain is a distributed ledger technology that ensures secure and transparent transactions.
It powers cryptocurrencies like Bitcoin and provides decentralized control."""

def preprocess(text):
    text = re.sub(r'[^\w\s]', '', text.lower())
    words = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    return [w for w in words if w not in stop_words]

tokens1 = preprocess(text1)
tokens2 = preprocess(text2)

set1, set2 = set(tokens1), set(tokens2)
jaccard_sim = len(set1 & set2) / len(set1 | set2)
print(f"Jaccard Similarity: {jaccard_sim:.3f}")

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([text1, text2])
cos_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
print(f"Cosine Similarity: {cos_sim:.3f}")

if cos_sim > jaccard_sim:
    print("Cosine Similarity provides a better insight as it captures term importance.")
else:
    print("Jaccard Similarity shows basic word overlap without considering term weight.")
```

### Q5. Write a short review for a product or service.
#### 1. Use TextBlob or VADER to find polarity & subjectivity for each review.
#### 2. Classify reviews into Positive / Negative / Neutral.
#### 3. Create a word cloud using the wordcloud library for all positive reviews.

```python
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt

reviews = [
    "The laptop battery life is amazing and performance is top-notch!",
    "The customer service was slow and unhelpful.",
    "It's an average product with decent features.",
    "Absolutely love the camera quality and sleek design!",
    "The software is buggy and keeps crashing."
]

sentiments = []
positive_words = ""

print("=== Sentiment Analysis ===")
for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity

    if polarity > 0:
        label = "Positive"
        positive_words += " " + review
    elif polarity < 0:
        label = "Negative"
    else:
        label = "Neutral"

    sentiments.append((review, polarity, subjectivity, label))
    print(f"Review: {review}\n→ Polarity: {polarity:.2f}, Subjectivity: {subjectivity:.2f}, Sentiment: {label}\n")

wc = WordCloud(width=800, height=400, background_color='white').generate(positive_words)

plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Positive Reviews", fontsize=14)
plt.show()
```

### Q6. Choose your own paragraph (~100 words) as training data.
#### 1. Tokenize text using Tokenizer() from keras.preprocessing.text
#### 2. Create input sequences and build a simple LSTM or Dense model
#### 3. Train the model and generate 2–3 new lines of text starting from any seed word you provide.

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, Flatten
from tensorflow.keras.preprocessing.sequence import pad_sequences

data = """ Generative AI is transforming industries with its ability to produce creative content.
It enhances productivity by automating tasks and supporting human innovation.
From writing code to generating images, GenAI expands possibilities.
It’s increasingly used in healthcare, education, and media.
With responsible use, this technology can amplify human potential and reshape the future."""

tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
word_index = tokenizer.word_index
total_words = len(word_index) + 1

input_sequences = []
tokens = tokenizer.texts_to_sequences([data])[0]

for i in range(2, len(tokens)):
    n_gram_seq = tokens[:i]
    input_sequences.append(n_gram_seq)

max_len = max(len(x) for x in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_len)

X = input_sequences[:, :-1]
y = input_sequences[:, -1]
y = to_categorical(y, num_classes=total_words)

model = Sequential()
model.add(Embedding(total_words, 10, input_length=max_len-1))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, verbose=0)

seed_text = "generative"
next_words = 5

for _ in range(next_words):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=max_len-1)
    predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)
    for word, index in tokenizer.word_index.items():
        if index == predicted:
            seed_text += " " + word
            break

print("\nGenerated text:")
print(seed_text)
```
