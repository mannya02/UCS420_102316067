### Q1.) Ask the user to enter a value (e.g., M) for a mathematical function.
(i) Generate x values from -10 to 10 using np.linspace().
(ii) Compute y values for:
(a) y= M⋅x^2
(b) y=M⋅sin(x)
(iii) Plot both functions in a single figure:
(a) Use different colors and line styles.
(b) Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt
M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10)
y1 = M * x**2
y2 = M * np.sin(x)
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label=r'$y = Mx^2$', color='b', linestyle='-')
plt.plot(x, y2, label=r'$y = M\sin(x)$', color='r', linestyle='--')
plt.legend()
plt.grid(True)
plt.title(f"Plot of Functions for M ={M} ")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### Q2.) Create a dataset of five subjects and scores.
(i) Convert it into a Pandas DataFrame.
(ii) Plot the scores using a Seaborn bar plot with:
(a) Different colors for each bar.
(b) Annotations on top of each bar.
(c) Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
data = {"Subject": ["Math", "Science", "English", "History", "Computer"],"Score": [80, 75, 66, 65, 99]}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 5))
graph = sns.barplot(x="Subject", y="Score", data=df)  
graph.bar_label(graph.containers[0], fmt='%d', fontsize=12, color='black', label_type='edge')
plt.title("Subject Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Q3.) Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().
(i) Create a 2x2 subplot layout displaying:
(a) Line plot showing cumulative sum.
(b) Scatter plot with random noise.
(ii) Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(102316067)
data = np.random.randn(50)
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
axs[0, 0].plot(np.cumsum(data), color='b')
axs[0, 0].set_title("Cumulative Sum")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Cumulative Value")
axs[0, 0].grid(True)
axs[0, 1].scatter(range(len(data)), data, color='r')
axs[0, 1].set_title("Scatter Plot with Noise")
axs[0, 1].set_xlabel("Index")
axs[0, 1].set_ylabel("Value")
axs[0, 1].grid(True)
axs[1, 0].axis('off')
axs[1, 1].axis('off')
plt.show()
```

### Q4.) Download Data-set from the below link https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv. Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

df = pd.read_csv("company_sales_data.csv")
sns.lineplot(x="month_number", y="total_profit", data=df, marker="o", color="b")
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

for col in df.columns[1:-1]:  
    sns.lineplot(x="month_number", y=col, data=df, label=col)
plt.title("Monthly Product Sales")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.legend()
plt.grid(True)
plt.show()

df.plot(kind='bar', figsize=(10, 5), alpha=0.75)
plt.title("Company Sales Data")
plt.xlabel("Month")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
