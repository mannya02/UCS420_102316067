## Q1) You are a data analyst at an e-commerce company. Your task is to analyze sales data,clean and process it, perform statistical analysis, and generate visual insights to help the business make informed decisions.

### Part I: Randomized Sales Data Generation (NumPy)
#### 1. Initialize Random Seed
#### 2. Generate Sales Data
### Create a NumPy array of random sales data for 12 months (shape = (12, 4)), where each column represents a product category:
#### Electronics, Clothing, Home & Kitchen, Sports.
#### Sales values should be between 1000 and 5000 units per month.
#### The range should be randomized using the seed.
### 3. Convert to DataFrame: Convert the cleaned NumPy array into a Pandas DataFrame with:
#### i. Columns: Electronics, Clothing, Home & Kitchen, Sports
#### ii. Index: Months from Jan to Dec

```python
import numpy as np
import pandas as pd

rno=102316067
np.random.seed(rno)
# Random sales between 1000 and 5000 for 12 months and 4 product categories
sales_data = np.random.randint(1000, 5001, size=(12, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, index=months, columns=columns)
print(df)
```

#### Total Sales: Sum of all categories per month.
#### Growth Rate: Percentage change in Total Sales from the previous month.
### 5. Apply a randomized discount:
#### If your roll number is even, apply a 10% discount to the Electronics category.
#### If odd, apply a 15% discount to the Clothing category.

```python
print("First 5 rows of the Dataframe:")
print(df.head())

print("Summary statistics of the DataFrame:")
print(df.describe())

total_per_cat = df.sum()
print("Total sales per category=",total_per_cat)

total_per_month = df.sum(axis=1)
print("Total sales per month=",total_per_month)

growth = df.pct_change().mean()
print("Average sales growth=",growth)
print("Total sales=")
df['Total Sales'] = df.sum(axis=1)
print("Growth Rate=")
df['Growth Rate (%)'] = df['Total Sales'].pct_change() * 100
df
print("Discount=")
if rno % 2 == 0:
    df['Electronics'] = df['Electronics'] * 0.9  # 10% discount
else:
    df['Clothing'] = df['Clothing'] * 0.85  # 15% discount
```

## Q2) For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
### i. Find element-wise absolute value
### ii. Find the 25th, 50th, and 75th percentile of flatened array, for each column, for each row.
### iii. Mean, Median and Standard Deviation of flatened array, of each column, and each row.

```python
arr = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Absolute Value
abs_arr = np.abs(arr)
print("Element wise absolute value",abs_arr)

# ii. Percentiles
print("Percentiles:")
flat = array.flatten()
p_flat = np.percentile(flat, [25, 50, 75])
p_col = np.percentile(array, [25, 50, 75], axis=0)
p_row = np.percentile(array, [25, 50, 75], axis=1)
print(p_flat)
print(p_col)
print(p_row)

# iii. Mean, Median, Std Dev
mean_flat = flat.mean()
print("Mean of flatened array",mean_flat)
median_flat = np.median(flat)
print("Median of flatened array",median_flat)
std_flat = flat.std()
print("Standard Deviation of flatened array",std_flat)

mean_col = array.mean(axis=0)
print("Mean of each column",mean_col)
median_col = np.median(array, axis=0)
print("Median of each column",median_col)
std_col = array.std(axis=0)
print("Standard Deviation of each column",std_col)

mean_row = array.mean(axis=1)
print("Mean of each row", mean_row)
median_row = np.median(array, axis=1)
print("Median of each row",median_row)
std_row = array.std(axis=1)
print("Standard Deviation of each row",std_row)
```

## Q4) WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
def swap_elements(lst, i, j):
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp

l = [10, 20, 30, 40, 50]
swap_elements(l, 1, 3)  
print("List after swapping:", l)
```

## Q5) Write a Python program to swap two elements in a set by converting it to a list first.

```python
def swap_elements_in_set(s, elem1, elem2):
    lst = list(s)
    if elem1 not in lst or elem2 not in lst:
        print("One or both elements not found in the set.")
        return s
    idx1 = lst.index(elem1)
    idx2 = lst.index(elem2)
    print(f"Before swapping: {lst}")
    lst[idx1], lst[idx2] = lst[idx2], lst[idx1]
    print(f"After swapping: {lst}")
    return set(lst)
s = {10, 20, 30, 40}
print("Original set:", s)
s = swap_elements_in_set(s, 20, 40)
```
